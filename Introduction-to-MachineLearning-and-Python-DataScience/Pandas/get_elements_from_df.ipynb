```python
import pandas as pd
import numpy as np
```

```python
#Create DataFrame from dicitonary
df = pd.DataFrame({
    'Col 1': {'Row 1': 1, 'Row 2': 2, 'Row 3': 3},
    'Col 2': {'Row 1': 4, 'Row 2': 5, 'Row 3': 6},
    'Col 3': {'Row 1': 7, 'Row 2': 8, 'Row 3': 9}
    })

print(df)
print()

#Add a column with a dictionary
df['Col 4'] = {'Row 1': 10, 'Row 2': 20, 'Row 3': 30}
print(df)
print()

#adding 4th row using 'loc' method
df.loc['Row 4'] = [40,50,60,70] #4 values because we now have 4 columns
print(df)
```

```python
#Slicing entire columns or rows from DataFrames
df_2 = pd.DataFrame({
    'Col 1': {'Row 1': 1, 'Row 2': 2, 'Row 3': 3, 'Row 4': 40},
    'Col 2': {'Row 1': 4, 'Row 2': 5, 'Row 3': 6, 'Row 4': 50},
    'Col 3': {'Row 1': 7, 'Row 2': 8, 'Row 3': 9, 'Row 4': 60},
    'Col 4': {'Row 1': 10, 'Row 2': 20, 'Row 3': 30, 'Row 4': 70}
    })

print(df_2)
print('-----------------------------------')
print('This is the first column as a Series:')

first_col = df_2['Col 1']#this is a Series
print(first_col)
print(type(first_col))
print('-----------------------------------')

print('This is the first row as a Series:')

first_row = df_2.loc['Row 1']#this is a Series. .loc expects a label
print(first_row)
print(type(first_row))
print('-----------------------------------')

print('This is the first row as a DataFrame:')

first_row_formatted = df_2.loc[['Row 1']]#this is a data frame
print(first_row_formatted)
print(type(first_row_formatted))
```

```python
#Getting elements from dataFrames
df_3 = pd.DataFrame({
    'Col1': {'Row 1': 1, 'Row 2': 2, 'Row 3': 3, 'Row 4': 40},
    'Col2': {'Row 1': 4, 'Row 2': 5, 'Row 3': 6, 'Row 4': 50},
    'Col3': {'Row 1': 7, 'Row 2': 8, 'Row 3': 9, 'Row 4': 60},
    'Col4': {'Row 1': 10, 'Row 2': 20, 'Row 3': 30, 'Row 4': 70}
    })

print('This is the first column:')

first_col_method = df_3.Col1
print(first_col_method)
print()
print('This is the second column:')

second_column = df_3.get('Col2')#you can use the 'get' method for columns only
print(second_column)

print()
print('This is the first row:')
first_row_method = df.loc['Row 1']#using 'loc' to get rows by label name. 'iloc' would be used to get by default, positional index value
print(first_row_method)

print()
print('These are individual elements:')

#Getting a single element
top_left = df_3['Col1'].loc['Row 1']
print(top_left)

bottom_right = df_3.loc['Row 4'].iloc[3]
print(bottom_right)

end_of_second_row = df_3.iloc[1,3]#row,columnformat
print(end_of_second_row)

```

```python
#Column Slicing
df_4 = pd.DataFrame({
    'Col1': {'Row 1': 1, 'Row 2': 2, 'Row 3': 3, 'Row 4': 40},
    'Col2': {'Row 1': 4, 'Row 2': 5, 'Row 3': 6, 'Row 4': 50},
    'Col3': {'Row 1': 7, 'Row 2': 8, 'Row 3': 9, 'Row 4': 60},
    'Col4': {'Row 1': 10, 'Row 2': 20, 'Row 3': 30, 'Row 4': 70}
    })


print("These are the first 2 cols from the DataFrame:")
print()

first_2_cols = df_4[['Col1', 'Col2']]
print(first_2_cols)

print()
print('This is the orignal df but rearranged:')

rearranged_df = df_4[['Col4','Col3', 'Col1', 'Col3']]
print(rearranged_df)
```

```python
#Row Slicing
df_5 = pd.DataFrame({
    'Col1': {'Row 1': 1, 'Row 2': 2, 'Row 3': 3, 'Row 4': 40},
    'Col2': {'Row 1': 4, 'Row 2': 5, 'Row 3': 6, 'Row 4': 50},
    'Col3': {'Row 1': 7, 'Row 2': 8, 'Row 3': 9, 'Row 4': 60},
    'Col4': {'Row 1': 10, 'Row 2': 20, 'Row 3': 30, 'Row 4': 70}
    })


print()
print('These are The first rows of the DataFrame')

first_2_rows = df_5.iloc[[0,1]]
print(first_2_rows)

print()
print('These are the last 2 rows')

last_2_rows = df_5.loc[['Row 3', 'Row 4']]
print(last_2_rows)

print()
print('This is the original df with rows shuffled:')

df_shuffled = df_5.loc[['Row 4', 'Row 1', 'Row 3', 'Row 2']]
print(df_shuffled)

print()
print('These are the first 2 rows of the new df:')

iloc_first_2 = df_shuffled.iloc[:2]#This grabs rows at index 0,1
print(iloc_first_2)

print()
print("these are the last 2 rows of new df:")

iloc_last_2 = df_shuffled.iloc[2:4]#4 is oob but since it's not included as per this syntax, there is no exception thrown
print(iloc_last_2)

print()
```

```python
#Advanced Slicing
df_6 = pd.DataFrame({
    'Col1': {'Row 1': 1, 'Row 2': 2, 'Row 3': 3, 'Row 4': 40},
    'Col2': {'Row 1': 4, 'Row 2': 5, 'Row 3': 6, 'Row 4': 50},
    'Col3': {'Row 1': 7, 'Row 2': 8, 'Row 3': 9, 'Row 4': 60},
    'Col4': {'Row 1': 10, 'Row 2': 20, 'Row 3': 30, 'Row 4': 70}
    })

print('This is the last two using -2:')
iloc_last_2_any_size = df_6[-2:]#not Pandas specific but works with Pandas DataFrames.
print(iloc_last_2_any_size)

print()
print("This is a slice skipping every other row:")
skiiping_by_2 = df_6[::2]# :: means we want every row and every column. the number is the step count
print(skiiping_by_2)

print()
print('This is all rows with select columns:')

iloc_first_2_columns = df_6.iloc[:,:2]
print(iloc_first_2_columns)

print()
print("This is first two rows and columns:")

top_left_4_items = df_6.iloc[:2,:2]
print(top_left_4_items)

print()
print('This is bool fetching:')
bool_fetching = df_6[df_6['Col1'] > 2 ]#this gets all rows that have a value in col1 that is greater than 2
print(bool_fetching)

print()
print('These are the first n rows:')
n_rows = df_6.head(2)#by default head method returns the first 5
print(n_rows)

print()
print('These are the last n rows:')
n_rows_tail = df_6.tail(2)#by default tail method returns the first 5
print(n_rows_tail)
```
