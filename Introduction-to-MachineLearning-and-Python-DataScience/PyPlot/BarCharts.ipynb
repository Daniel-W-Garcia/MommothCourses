```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Bar Chart
bar_chart_data = {'Army': 5, 'Navy': 1, 'Marines': 4, 'Airforce': 10, 'Coasties': 2 }

bars_plot = plt.bar(bar_chart_data.keys(), bar_chart_data.values(), width=.5)

plt.setp(bars_plot, color='b', edgecolor='black' )
plt.xlabel('Branch')
plt.ylabel('Job Satisfaction')
plt.title('Armed Forces Job Satisfaction')

plt.show()
```
