```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Create 3D Plot
x = np.arange(1,101)
y = np.random.randint(1, 101, size=100)
z = np.random.randint(1, 101, size=100)

axes = plt.axes(projection='3d')
axes.scatter3D(x,y,z)
axes.set_xlabel('This is X Label')
axes.set_ylabel('This is Y Label')
axes.set_zlabel('This is Z Label')
plt.show()
```
